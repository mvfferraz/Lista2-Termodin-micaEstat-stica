```python
import numpy as np

def simpson(f,a,b,N=1000):
    dx = (b-a)/N
    x = np.linspace(a,b,N+1)
    y = f(x)
    S = dx/3 * np.sum(y[0:-1:2] + 4*y[1::2] + y[2::2])
    return S
```

```python
simpson( lambda x: ((0.083*350)/(x-2.6738*10**-2)-(15.551)/(350**(1/2)*x*(x+2.6738*10**-2))),5,1,1000)
```

```python
#Cálculo de p em função de ro

import matplotlib.pyplot as plt
import numpy as np
ro = np.arange(0,12.3, 0.3)
p = ((33.258*ro)-(7.5885*ro**2)+(1.0306*ro**3)-(0.058757*ro**4)-(0.0033566*ro**5)+(0.00060696*ro**6))

#Cálculo do volume (=1/ro)

v = (1/ro)
```

```python
#Plot vdW
vdw = ((0.0832*350)/(v-0.090494))-(9.3919/v**2)
plt.plot(ro,vdw, 'g')
plt.xlabel("p(/bar)", size=12)
plt.ylabel(r'$\rho$ (/mol/L)', size=12)
plt.show()
```

```python
#Plot RK
rk = ((0.0832*350)/(v-0.062723))-((183.02)/((350**(1/2))*v*(v+0.062723)))
plt.xlabel("p(/bar)", size=12)
plt.ylabel(r'$\rho$ (/mol/L)', size=12)
plt.plot(ro,rk, 'g')
plt.show()

```

```python
#Comparação vdW x RK
plt.xlabel("p(/bar)", size=12)
plt.ylabel(r'$\rho$ (/mol/L)', size=12)
plt.plot(ro, rk, 'r', label='RK')
plt.plot(ro, vdw, 'b', label='vdW')  
plt.plot(ro, p, 'o', label='Experimental') 
plt.legend(loc="lower left")
plt.show()
```

# Questão 6

```python
#Lennard-Jones para e=5 kJ.mol-1 e sigma = 1.5nm (1)

#Termo atrativo

import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1.3,4.5, 0.01)
at = (4*5*((1.5)**(12)/x**12))

#Termo repulsivo
x = np.arange(1.3,4.5, 0.01)
rep = -(4*5*((1.5)**(6)/x**6))

#Potencial
lj1 = (4*5*((1.5)**(12)/x**12))-(4*5*((1.5)**(6)/x**6))

#Plot
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.xlabel("r (nm)", size=12)
fig.gca().set_ylabel('V (kJ/mol)', size=12)
plt.plot(x, at, 'r', label='Atrativo')
plt.plot(x, rep, 'b', label='Repulsivo')  
plt.plot(x, lj1, 'g', label='Lennard-Jones') 
plt.legend(loc="upper right")
plt.show()
```

```python
#Lennard-Jones para e=10 kJ.mol-1 e sigma = 1.4nm (2)

#Termo atrativo

import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1.3,4.5, 0.01)
at2 = (4*10*((1.4)**(12)/x**12))

#Termo repulsivo
x = np.arange(1.3,4.5, 0.01)
rep2 = -(4*10*((1.4)**(6)/x**6))

#Potencial
lj2 = (4*10*((1.4)**(12)/x**12))-(4*10*((1.4)**(6)/x**6))

#Plot
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.xlabel("r (nm)", size=12)
fig.gca().set_ylabel('V (kJ/mol)', size=12)
plt.plot(x, at2, 'r', label='Atrativo')
plt.plot(x, rep2, 'b', label='Repulsivo')  
plt.plot(x, lj2, 'g', label='Lennard-Jones') 
plt.legend(loc="upper right")
plt.show()
```

```python
#Comparativo potencial de LJ para situações (1) e (2)
plt.plot(x, lj1, 'b', label='e=5 kJ/mol, sigma=1.5 nm')  
plt.plot(x, lj2, 'r', label='e=10 kJ/mol, sigma=1.4 nm') 
plt.legend(loc="upper right")
plt.xlabel("r (nm)", size=12)
plt.ylabel('V (kJ/mol)', size=12)

```
